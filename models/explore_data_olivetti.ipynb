```python
from sklearn import datasets

data = datasets.fetch_olivetti_faces()
```

## Double check the target data

Targets should be all the same person

```python
figsize(30, 120)
fig, axes = plt.subplots(40, 10)

for target_ind in range(40):
    faces = data.images[data['target'] == target_ind]
    for i in range(10):
        plt.subplot(40,10, 10*target_ind+i+1)
        plt.xticks(visible=False)
        plt.yticks(visible=False)
    
        plt.imshow(faces[i], cmap='grey')
        
```
